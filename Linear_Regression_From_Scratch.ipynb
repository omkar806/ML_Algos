```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/placement.csv')
df
```

```python
new_cols = {'cgpa':'x' , 'package':'y'}
df = df.rename(columns=new_cols)
```

```python
#defining a loss funtion
def loss_fn(m,b,points):
  total_error = 0
  for i in range(len(points)):
    x=points.iloc[i].x
    y=points.iloc[i].y
    total_error += (y-(m*x+b))**2
  total_error=total_error/float(len(points))
```

```python
def gradient_descent(m_now,b_now,points,L):
  m_gradient = 0
  b_gradient = 0
  n=len(points)

  for i in range(n):
    x=points.iloc[i].x
    y=points.iloc[i].y
    m_gradient += -(2/float(n))*(x)*( y - (m_now*x + b_now))
    b_gradient += -(2/float(n))*( y - (m_now*x+b_now))
  m=m_now - L*m_gradient
  b=b_now - L*b_gradient

  return m,b

```

```python
plt.scatter(df.x,df.y,color='black')
```

```python
m=0
b=0
L = 0.0001
epochs = 200

for i in range(epochs):
  if i%50 == 0 :
    print(f'Epoch : {i}')
  m,b = gradient_descent(m,b,df,L)

print(m,b)
# plt.scatter(df.x,df.y,color='black')
# plt.plot(list(range(20,80)),[m*x + b for x in range(20,80)],color='red')
# plt.show()
```
